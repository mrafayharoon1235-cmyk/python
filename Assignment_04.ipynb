Q3 — Loop Through Dictionary
person = {

"city": "Karachi"
}
Print keys and values using loop.


```python
person = {
    "name": "Ali",
    "age": 25,
    "city": "Karachi"
}

# Loop through dictionary
for key, value in person.items():
    print(key, ":", value)
```

Q4 — Count Frequency of Numbers
numbers = [1,2,2,3,3,3,4,4,4,4]
Create frequency dictionary.


```python
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

frequency = {}

for num in numbers:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1

print(frequency)
```

Q5 — Word Counter
Take sentence input and count word frequency.


```python
sentence = input("Enter a sentence: ")
words = sentence.split()

frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Print result
print(frequency)
```

Q6 — Student Marks & Average
marks = {"Math":80, "English":70, "Physics":90}
Calculate total and average.


```python
marks = {"Math": 80, "English": 70, "Physics": 90}
total = 0
for value in marks.values():
    total += value

average = total / len(marks)

print("Total Marks:", total)
print("Average Marks:", average)
```

Q7 — Highest Scoring Student
students = {
"Ali":85,
"Sara":92,
"Ahmed":78,
"Zara":95
}
Find highest scoring student.


```python
students = {
    "Ali": 85,
    "Sara": 92,
    "Ahmed": 78,
    "Zara": 95
}

highest_student = ""
highest_marks = 0

for name, marks in students.items():
    if marks > highest_marks:
        highest_marks = marks
        highest_student = name

print("Highest Scoring Student:", highest_student)
print("Marks:", highest_marks)
```

Q8 — Merge Dictionaries
dict1 = {"a":1,"b":2}

```python
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}

merged_dict = dict1.copy()
merged_dict.update(dict2)

print(merged_dict)
```
