# QUESTION 1: Working with Lists

1.1.	Create a list called numbers containing at least 8 integers.

```python
numbers = [5, 12, 7, 20, 3, 15, 9, 18]
```

1.2.	Using a for loop, do the following: 
o	Print all numbers in the list.
o	Print the sum of all numbers.
o	Print only the even numbers.
Requirements:
•	Use at least one loop.
•	Use conditional statements.
•	Display clear output.


```python
numbers = [5, 12, 7, 20, 3, 15, 9, 18]

total = 0

print("All numbers in the list:")
for num in numbers:
    print(num)
    total += num

print("\nSum of all numbers:", total)

print("\nEven numbers in the list:")
for num in numbers:
    if num % 2 == 0:
        print(num)
```

# QUESTION 2: Dictionary Basics

2.1.	Create a dictionary called student with the following keys:
o	name
o	age
o	course
o	marks (must store a list of 3 marks)


```python
student = {
    "name": "Ali",
    "age": 20,
    "course": "Python",
    "marks": [85, 90, 78]
}
```

2.2.	Using loops:
o	Print all keys and their values.
o	Calculate the average of the marks.
o	Display whether the student Passed (average ≥ 50) or Failed.
Requirements:
•	Dictionary must contain a list.
•	Use loops to calculate the average.


```python
student = {
    "name": "Ali",
    "age": 20,
    "course": "Python",
    "marks": [85, 60, 75]
}

print("Student Information:")

for key, value in student.items():
    print(key, ":", value)

total = 0
count = 0

for mark in student["marks"]:
    total += mark
    count += 1

average = total / count

print("\nAverage Marks:", average)

if average >= 50:
    print("Result: Passed")
else:
    print("Result: Failed")
```

# QUESTION 3: List of Dictionaries

3.1.	Create a list called employees.
3.2.	Each employee must be stored as a dictionary containing:
o	name
o	department
o	salary
3.3.	Using loops:
o	Print details of all employees.
o	Find the employee with the highest salary.
o	Calculate the total salary expense.
Requirements:
•	Minimum 3 employees.
•	Use comparison logic.
•	Use loops to process data.


```python
employees = [
    {"name": "Ali", "department": "IT", "salary": 50000},
    {"name": "Sara", "department": "HR", "salary": 60000},
    {"name": "Ahmed", "department": "Finance", "salary": 55000}
]

print("Employee Details:\n")

for emp in employees:
    print("Name:", emp["name"])
    print("Department:", emp["department"])
    print("Salary:", emp["salary"])
    print()

highest_salary = employees[0]["salary"]
highest_employee = employees[0]["name"]

for emp in employees:
    if emp["salary"] > highest_salary:
        highest_salary = emp["salary"]
        highest_employee = emp["name"]

print("Employee with Highest Salary:", highest_employee, "-", highest_salary)

total_salary = 0

for emp in employees:
    total_salary += emp["salary"]

print("Total Salary Expense:", total_salary)
```

# QUESTION 4: While Loop with User Input

1.	Ask the user to enter numbers.
2.	Store the numbers in a list.
3.	Stop when the user enters -1.
4.	After the loop ends:
o	Print the complete list.
o	Print the largest number.
o	Print the smallest number.
Requirements:
•	Must use a while loop.
•	Must store input in a list.
•	Handle user input correctly.


```python
numbers = []

while True:
    num = int(input("Enter a number (-1 to stop): "))
    
    if num == -1:
        break
    
    numbers.append(num)

print("\nNumbers entered:", numbers)

largest = numbers[0]
smallest = numbers[0]

for n in numbers:
    if n > largest:
        largest = n
    if n < smallest:
        smallest = n

print("Largest number:", largest)
print("Smallest number:", smallest)
```

# QUESTION 5: Word Frequency Counter

1.	Ask the user to enter a sentence.
2.	Convert the sentence into a list of words.
3.	Use a dictionary to count how many times each word appears.
4.	Print the word frequency clearly.
Example Format (for understanding only):
apple: 3
banana: 2
orange: 1
Requirements:
•	Use string splitting.
•	Use a dictionary for counting.
•	Use loops for processing.


```python

sentence = input("Enter a sentence: ")


words = sentence.split()

word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print("\nWord Frequency:")
for word, count in word_count.items():
    print(word + ":", count)
```

# QUESTION 6: Nested Loops – Multiplication Table

Using nested loops, print a multiplication table from 1 to 5.
Expected Output Format:
1 2 3 4 5
2 4 6 8 10
3 6 9 12 15
4 8 12 16 20
5 10 15 20 25
Requirements:
•	Must use nested loops.
•	Proper formatting required.

General Requirements
•	Use:
o	At least 2 lists
o	At least 2 dictionaries
o	At least 3 loops
o	At least 1 nested loop
•	Include comments explaining your code.
•	Proper indentation is required.
•	Code must run without errors.
•	Submit your .py file before the due date.



```python

numbers = [1, 2, 3, 4, 5]

departments = ["IT", "HR", "Finance"]



student = {
    "name": "Ali",
    "age": 20,
    "course": "Python"
}

employee_salary = {
    "Ali": 50000,
    "Sara": 60000,
    "Ahmed": 55000
}


print("Student Information:")
for key, value in student.items():
    print(key, ":", value)

print()


print("Employee Salaries:")
for name, salary in employee_salary.items():
    print(name, "earns", salary)

print()


print("Departments:")
for dept in departments:
    print(dept)

print()


print("Multiplication Table (1 to 5):")

for i in numbers:              # Outer loop
    for j in numbers:          # Inner loop
        print(i * j, end=" ")  # Print in same line
    print()                    # Move to next line after each row
```
